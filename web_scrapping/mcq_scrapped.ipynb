```python
from tqdm import tqdm
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url='https://folderat.com/Reference/1649/%D8%A3%D8%B3%D8%A6%D9%84%D8%A9-%D8%AF%D9%8A%D9%86%D9%8A%D8%A9-%D9%85%D8%B9-%D8%AE%D9%8A%D8%A7%D8%B1%D8%A7%D8%AA'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
scrap = soup.find_all('div', attrs={'class':'question'})
# scrap
```

```python
questions = []
answers = []
for i in scrap:
    questions.append(i.find('li').text)
    answers.append(i.find('strong').text)
```

```python
pages = soup.find_all('div', attrs={'class': "w3-display-container"})
pages
```

```python
links = []
pages = soup.find_all('div', attrs={'class': "w3-display-container"})

for page in tqdm(pages):
    link = page.find('a').get('href')
    links.append(link)

    
len(links)
```

```python
questions = []
answers = []


for i in tqdm(links):
    response = requests.get(i)
    soup = BeautifulSoup(response.text, 'html.parser')
    scrap = soup.find_all('div', attrs={'class':'question'})

    for i in scrap:
        questions.append(i.find('li').text)
        answers.append(i.find('strong').text)
    
```

```python
df = pd.DataFrame({'Questions':questions, 'Answer':answers})
df
```
