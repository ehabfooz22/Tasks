```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
import csv
```

```python
sources = requests.get("https://folderat.com/Reference/1649/%D8%A3%D8%B3%D8%A6%D9%84%D8%A9-%D8%AF%D9%8A%D9%86%D9%8A%D8%A9-%D9%85%D8%B9-%D8%AE%D9%8A%D8%A7%D8%B1%D8%A7%D8%AA").text
sources
```

```python
soup = BeautifulSoup(sources , 'lxml')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
article = soup.find('div', class_ = "question")
print(article)
```

```python
match = soup.title.text
print(match)
```

```python
articles = soup.find_all('div', class_ = "question")
for article in articles:
    q =  article.ul.li.text
    print(q)
```

```python
links = []
articles = soup.find_all('div', class_ = "w3-display-container")
for article in articles:
    link =article.a.get('href')
    links.append(link)
   
```

```python
links[0]
```

```python
len(links)
```

```python
for i in links :
    x = requests.get(i).text
    soup = BeautifulSoup(x , 'lxml')
print(soup)
```

```python
ques= []
for i in links :
    x = requests.get(i).text
    soup = BeautifulSoup(x , 'lxml')
    for i in soup:
        articles = soup.find_all('div', class_ = "question")
        for article in articles:
            q =  article.ul.li.text
            ques.append(q)
```
